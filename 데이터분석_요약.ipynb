## *   판다스
    
*   파이썬에서의 가장 대표적인 데이터분석 도구
    
*   표 형태의 데이터를 다루는 다양한 함수를 제공하고 있음

*   표 형태의 데이터를 데이터프레임이라고 하며, 각각의 독립적인 열을 시리즈라고 부름


<br><br>



##*   판다스 열 접근
*   df[열]
*   df[[열1,열2]]



<br><br>



##*   판다스 행,열 접근

*   loc: 이름으로 행과 열에 접근
    *   loc[행,열]
*   iloc: 순서로 행과 열에 접근
    *   iloc[행,열]
<br><br>



##*   판다스의 to_datetime

*   pd.to_datetime() 함수를 통해서 문자열로 된 날짜 데이터를 날짜 데이터로 변환 가능



<br><br>



##*   판다스의 groupby 함수

*   gropuby 함수를 통해 데이터프레임 내에 있는 데이터를 내가 원하는 대로 집단화할 수 있음


<br><br>
##*   다양한 데이터 시각화 라이브러리

*   matplotlib,seaborn,pandas의 시각화 함수를 통해서 데이터프레임의 데이터를 쉽게 시각화

*   matplotlib 라이브러리는 커스터마이징 옵션이 많아, 내가 원하는 형태로 그림 그리기가 원활함

*   seaborn 라이브러리는 자동으로 예쁜 그림을 그려주나, 커스터마이징 크게 자유롭지 않음

*   pandas 안에 있는 가시화함수로 간단한 그림을 그릴 수 있음


<br><br>
##* pandas의 concat
*   여러개의 데이터프레임을 하나의 데이터프레임으로 합칠 수 있음

*   매개변수 axis는 축을 의미함(1은 열방향, 0은 행방향)


<br><br>
##* pandas의 merge
*   merge 함수는 두 개의 데이터 프레임에서 각각 기준열을 정하고, 그 기준열에 있는 데이터에 맞게 데이터프레임을 합쳐주는 역할을 함

*   df1.merge(df2,left_on=df1의 기준열, right_on=df2의 기준열)

*   만약 두 데이터프레임의 기준열의 이름이 같다면 left_on,right_on 대신 on으로 대체해도 된다.


<br><br>
##* 누락값
*   누락값은 어떤것과도 비교를 해도 False가 나오기 떄문에 if문을 사용해서 누락값인지 확인을 할 수 없다. 누락값을 확인하려면 pd.isnull() 함수로 값이 누락값인지 아닌지 확인할 수 있다.

*   from numpy import NaN,nan,NAN에서 누락값을 제공하기 때문에 누락값을 생성할 수도 있다.

*   df.fillna() 함수를 통해 누락값을 채울 수 있다

<br><br>
##*  pandas의 melt
*   melt 함수는 넚은 데이터에서 같은 의미를 가진 여러 개의 하나의 열로 통합해준다

<br><br>
##*  pandas의 pivot_table
*   melt 함수와는 반대로 행에 있는 값을 열로 올릴 때 사용한다


<br><br>
##*  pandas의 rank
*   rank 함수는 특정 열을 기준으로 값에 순위를 매길 때 사용한다

*   다양한 method(average,min,max,first,dense 등) 함수로 내가 원하는 대로 순위를 조절하기가 편하다

<br><br>
##*  상관분석
*   상관분석은 두 변수 사이에 선형관계를 가지는지 분석하는 방법

*   상관관계는 -1~ 1사이의 값을 가지며 -1 또는 1에 가까울수록 높은 상관관계를 가진다.

*   scipy패키지의 stats 함수를 사용하면 쉽게 상관계수를 구할 수 있다

<br><br>
##*  정규표현식과 re 라이브러리
