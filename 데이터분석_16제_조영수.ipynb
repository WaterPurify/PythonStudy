문제 14) 타이타닉호 데이터프레임에서 생존에 가장 상관관계가 높은 열을 찾아보세요.
* "생존" 열은 survived 열입니다.

import pandas as pd
import seaborn as sns

titanic = pd.read_csv("titanic.csv")

```python
import scipy.stats as stats
survived = titanic['survived']
#pclass		age	sibsp	parch	fare
# sex embarked	class	who	adult_male	deck	embark_town	alive	alone
#위 클래스는 숫자가 아니라서 상과관계 표현 불가
pclass = titanic['pclass']
age = titanic['age']
sibsp = titanic['sibsp']
parch = titanic['parch']
fare = titanic['fare']

titanic['age'] = titanic['age'].fillna(titanic['age'].mean())

Result=[]

Pclass_survived = stats.pearsonr(survived,pclass)
Result.append(Pclass_survived)

age_survived = stats.pearsonr(survived,age)
Result.append(age_survived)

sibsp_survived = stats.pearsonr(survived,sibsp)
Result.append(sibsp_survived)

parch_survived = stats.pearsonr(survived,parch)
Result.append(parch_survived)

fare_survived = stats.pearsonr(survived,fare)
Result.append(fare_survived)


print(Result)
#따라서 p-class가 가장 상관관계가 높은 열입니다 (0.34)
print('{}가 가장 상관관계가 높은 열입니다'.format('p-class'))
```
